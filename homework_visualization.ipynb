```python
import pandas as pd

# Пути к загруженным файлам
train_path = 'train_data.csv'
test_path = 'test_data.csv'

# Загрузка данных
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# Просмотр первых строк тренировочного и тестового датасетов
# train_data.head(), test_data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Подготовка данных
X_train = train_data.drop(columns = ['Unnamed: 0', 'PassengerId', 'Survived'])
y_train = train_data['Survived']

# Создание и обучение модели дерева решений
clf = DecisionTreeClassifier(max_depth = 3, random_state = 21)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20, 10))
plot_tree(clf,
          feature_names = X_train.columns,
          class_names=['Не выжил', 'Выжил'],
          filled=True)
plt.title('Визуализация дерева решений')
plt.show()
```
